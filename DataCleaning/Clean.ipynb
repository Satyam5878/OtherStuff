## Import Statements:

```python
import pandas as pd
import datetime
s = "\nDone(*_*)"
print(s)
```

## Loading data:

```python
data = pd.read_csv("data.csv",header=2)
print(data.shape)
print(s)
```

## Converting Columns (Hour and Minute) to HH:MM:SS

```python
"""
Taking Column Hour and Minute and concatenating to form DateTime obj.
Then Series.dt.time to get time from DateTime Object.
And then Appending new column in data with column name "Time".
"""
data["Time"] = pd.to_datetime(data["Hour"].map(str)+":"+data["Minute"].map(str)).dt.time
print(s)
```


## Filtering rows:

```python
"""
t1 and t2 are time for filtering.
filter based and these times seperately and then concatenate the these dataframe to obtain the result.
"""
t1 = datetime.time(6,0,0)
t2 = datetime.time(19,0,0)
print(t1,t2)

data_i = pd.DataFrame(data[data["Time"] >= t1])
data_filter = data_i[data_i["Time"] < t2]

print(data_filter.shape)

data_filter.

print(s)
```
